```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np

# Style
CATEGORICAL_PALETTE = sns.color_palette("colorblind")
PRIMARY_COLOR = 'darkslategray'
FONT_FAMILY = 'sans-serif'

def set_readable_style():
    """
    Sets a consistent, readable, and accessible style for all plots.
    """
    sns.set_style("whitegrid")
    plt.rcParams.update({
        'figure.figsize': (14, 8),
        'axes.titlesize': 24,
        'axes.labelsize': 18,
        'xtick.labelsize': 14,
        'ytick.labelsize': 14,
        'legend.fontsize': 14,
        'font.family': FONT_FAMILY,
        'axes.titleweight': 'bold',
        'axes.titlepad': 20,
        'axes.labelpad': 15,
        'axes.titlelocation': 'center',
        'axes.grid': True,
        'grid.color': '#CCCCCC',
        'axes.edgecolor': 'black',
        'xtick.color': 'black',
        'ytick.color': 'black',
        'axes.labelcolor': 'black',
        'text.color': 'black'
    })

def load_data(filepath):
    """Loads data from a CSV file."""
    print(f"Loading data from {filepath}...")
    try:
        df = pd.read_csv(filepath)
        print("Data loaded successfully.")
        return df
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return None

def initial_exploration(df):
    """Performs initial data exploration."""
    if df is None: return
    print("\n--- Initial Data Exploration ---")
    print("\nDataFrame Info:")
    df.info()
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("\nDescriptive Statistics:")
    print(df.describe())
    print("--------------------------------\n")

def main():
    input_filename = 'shopping_behavior_updated.csv'
    set_readable_style()

    df = load_data(input_filename)
    if df is not None:
        initial_exploration(df)

if __name__ == '__main__':
    main()

```

```python
# Customer demographics
def analyze_customer_demographics(df, output_dir='visualizations'):
 
    if df is None: return
    print("Analyzing customer demographics...")
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle('Customer Demographics', fontsize=28, weight='bold', y=1.02)

    # Age Distribution
    sns.histplot(df['Age'], bins=25, kde=True, ax=axes[0], color=PRIMARY_COLOR, edgecolor='black')
    axes[0].set_title('Age Distribution')
    axes[0].set_xlabel('Age')
    axes[0].set_ylabel('Number of Customers')

    # Gender Distribution
    sns.countplot(x='Gender', data=df, ax=axes[1], palette=CATEGORICAL_PALETTE, order=df['Gender'].value_counts().index, edgecolor='black')
    axes[1].set_title('Gender Distribution')
    axes[1].set_xlabel('Gender')
    axes[1].set_ylabel('Number of Customers')
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(os.path.join(output_dir, 'customer_demographics.png'))
    print("Saved 'customer_demographics.png'")
    plt.show()

# Purchase behavior
def analyze_purchases(df, output_dir='visualizations'):
 
    if df is None: return
    print("Analyzing purchase behavior...")
    
    # Purchase Amount Distribution
    plt.figure(figsize=(14, 7))
    sns.histplot(df['Purchase Amount (USD)'], bins=30, kde=True, color=PRIMARY_COLOR, edgecolor='black')
    plt.title('Distribution of Purchase Amounts')
    plt.xlabel('Purchase Amount (USD)')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'purchase_amount_distribution.png'))
    print("Saved 'purchase_amount_distribution.png'")
    plt.show()

    # Product Categories by Purchase Count
    plt.figure(figsize=(14, 8))
    sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index, palette=CATEGORICAL_PALETTE, edgecolor='black')
    plt.title('Most Purchased Product Categories')
    plt.xlabel('Number of Purchases')
    plt.ylabel('Category')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'category_popularity.png'))
    print("Saved 'category_popularity.png'")
    plt.show()
    
    # Total Purchase Amount by Product Category
    category_spending = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 12))
    colors = CATEGORICAL_PALETTE
    explode = [0.05] + [0] * (len(category_spending) - 1)

    wedges, texts, autotexts = plt.pie(category_spending, labels=category_spending.index, autopct='%1.1f%%',
            colors=colors, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
            pctdistance=0.85, explode=explode, startangle=90)
    plt.setp(autotexts, size=16, weight="bold", color="black")
    plt.setp(texts, size=14)

    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)

    plt.title('Share of Total Spending by Category')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'category_purchase_share_pie.png'))
    print("Saved 'category_purchase_share_pie.png'")
    plt.show()
# Payment method trends 
def analyze_trends(df, output_dir='visualizations'):
    if df is None: return
    print("Analyzing seasonal and payment trends...")
    
    fig, axes = plt.subplots(1, 2, figsize=(20, 9))
    fig.suptitle('Seasonal and Payment Trends', fontsize=28, weight='bold', y=1.02)
    
    # Seasonal Purchases
    sns.countplot(x='Season', data=df, order=['Spring', 'Summer', 'Fall', 'Winter'], palette=CATEGORICAL_PALETTE, ax=axes[0], edgecolor='black')
    axes[0].set_title('Purchases by Season')
    axes[0].set_xlabel('Season')
    axes[0].set_ylabel('Number of Purchases')

    # Payment Method Popularity
    sns.countplot(y='Payment Method', data=df, order=df['Payment Method'].value_counts().index, palette=CATEGORICAL_PALETTE, ax=axes[1], edgecolor='black')
    axes[1].set_title('Preferred Payment Methods')
    axes[1].set_xlabel('Number of Purchases')
    axes[1].set_ylabel('Payment Method')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(os.path.join(output_dir, 'seasonal_and_payment_trends.png'))
    print("Saved 'seasonal_and_payment_trends.png'")
    plt.show()

def analyze_engagement(df, output_dir='visualizations'):
    """Analyzes customer engagement metrics with a focus on readability."""
    if df is None: return
    print("Analyzing customer engagement...")
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle('Customer Satisfaction and Engagement', fontsize=28, weight='bold', y=1.02)

    # Review Rating Distribution
    sns.histplot(df['Review Rating'], bins=20, kde=True, color=PRIMARY_COLOR, edgecolor='black', ax=axes[0])
    axes[0].set_title('Distribution of Review Ratings')
    axes[0].set_xlabel('Review Rating')
    axes[0].set_ylabel('Frequency')

    # Subscription Status
    subscription_counts = df['Subscription Status'].value_counts()
    colors = [CATEGORICAL_PALETTE[0], CATEGORICAL_PALETTE[1]]
    wedges, texts, autotexts = axes[1].pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', 
                colors=colors, startangle=90, wedgeprops=dict(width=0.4, edgecolor='black', linewidth=1.5))
    axes[1].set_title('Subscription Status')
    plt.setp(autotexts, size=16, weight="bold", color="black")
    plt.setp(texts, size=14)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(os.path.join(output_dir, 'customer_engagement.png'))
    print("Saved 'customer_engagement.png'")
    plt.show()
    
def main():
    input_filename = 'shopping_behavior_updated.csv'
    output_directory = 'visualizations'

    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
        print(f"Created directory: {output_directory}")

    set_readable_style()

    df = load_data(input_filename)
    if df is not None:
        analyze_customer_demographics(df, output_directory)
        analyze_purchases(df, output_directory)
        analyze_trends(df, output_directory)
        analyze_engagement(df, output_directory)
        print("\nAnalysis complete. Visualizations saved to the 'visualizations' directory.")

if __name__ == '__main__':
    main()
```
